```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('data/BBCNews.csv')
```

```python
df
```

Date Parsing


```python
today = datetime.today().date()

for index in range(len(df['publisheddate'])):
    date = df['publisheddate'][index]
    if date.endswith('ago'):
        if 'hour' in date:
            df.loc[index]['publisheddate'] = today.strftime('%d %B %Y')
        if 'days' in date:
            df.loc[index]['publisheddate'] = (today - timedelta(days=int(date.split(' ')[0]))).strftime('%d %B %Y')
    elif date[len(date)-4:].isnumeric() == False:
            df.loc[index]['publisheddate'] = date + ' 2023'
        
df['publisheddate'] = pd.to_datetime(df['publisheddate'])
```

Sorting by Date

```python
df = df.sort_values(by='publisheddate',ascending=False)
df
```

Removal of Sport Section

```python
df.section.value_counts()
```

```python
df = df[df['section'] != 'Sport']
```

Reset Index

```python
df = df.reset_index(drop=True)
df
```

```python
df.to_csv('BBCNewsProcessed.csv',index=False)
```
