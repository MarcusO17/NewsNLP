```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
import pandas as pd
```

```python
df = pd.read_csv('data/BBCNewsProcessed.csv')
```

```python
def getSentiment(polarity_score):
    neutrality_threshold = 0.2
    sentiment = polarity_score['compound']
    if sentiment < neutrality_threshold :
        return 'Positive'
    elif sentiment > neutrality_threshold :
        return 'Negative'
    else:
        return 'Neutral'
```

```python
sia = SIA()
for index in range(len(df)):
    pol_score = sia.polarity_scores(df['title'][index])
    df.loc[index,'sentiment'] = getSentiment(pol_score)
```
